```python
import pandas as pd
import numpy as np

df_GM = pd.read_excel('gross margin.xlsx', index_col = '3 months ending').drop("Unnamed: 0", axis = 1)
df_OM = pd.read_excel('operatingmargin.xlsx', index_col = '3 months ending').drop("Unnamed: 0", axis = 1)
df_PM = pd.read_excel('profit margin.xlsx', index_col = '3 months ending').drop("Unnamed: 0", axis = 1)
df_RA = pd.read_excel('return on asset.xlsx', index_col = '3 months ending').drop("Unnamed: 0", axis = 1)
df_RE = pd.read_excel('return on equity.xlsx', index_col = '3 months ending').drop("Unnamed: 0", axis = 1)
```

```python
import matplotlib.pyplot as plt
df_GM = df_GM.replace("", np.NaN).backfill()
index_list = []
for index in df_GM.columns:
    index_list.append(index)
plt.figure(figsize = (20,10))
plt.xlabel("Time by Quarters")
plt.ylabel("Gross Margin")
plt.ylim(0,80)
plt.plot(df_GM)
plt.legend(index_list,loc=2)
```

```python
df_OM = df_OM.replace("", np.NaN).backfill()
plt.figure(figsize = (20,10))
plt.xlabel("Time by Quarters")
plt.ylabel("Operating Margin")
plt.ylim(-5,40)
plt.plot(df_OM)
plt.legend(index_list,loc=2)
```

```python
df_PM = df_PM.replace("", np.NaN).backfill()
plt.figure(figsize = (20,10))
plt.xlabel("Time by Quarters")
plt.ylabel("Profit Margin")
plt.ylim(-5,30)
plt.plot(df_PM)
plt.legend(index_list,loc=2)
```

```python
df_RA = df_RA.replace("", np.NaN).backfill()
plt.figure(figsize = (20,10))
plt.xlabel("Time by Quarters")
plt.ylabel("Return on Asset")
plt.ylim(0,15)
plt.plot(df_RA)
plt.legend(index_list,loc=2)
```

```python
df_RE = df_RE.replace("", np.NaN).backfill()
plt.figure(figsize = (20,10))
plt.xlabel("Time by Quarters")
plt.ylabel("Retrun on Equity")
plt.ylim(0,40)
plt.plot(df_RE)
plt.legend(index_list,loc=2)
```
