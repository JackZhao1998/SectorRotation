```python
import pandas as pd
import numpy as np

df_PE = pd.read_excel('/Users/Shuoyiyang/Desktop/sector rotation/Fundamentals/PE_Ratio.xlsx', index_col = '3 Months Ending').drop("Unnamed: 0", axis = 1)
df_PB = pd.read_excel('/Users/Shuoyiyang/Desktop/sector rotation/Fundamentals/PB_Ratio.xlsx', index_col = '3 Months Ending').drop("Unnamed: 0", axis = 1)
df_EV2Sales = pd.read_excel('/Users/Shuoyiyang/Desktop/sector rotation/Fundamentals/EV2Sales.xlsx', index_col = '3 Months Ending').drop("Unnamed: 0", axis = 1)
df_EV2EBIT = pd.read_excel('/Users/Shuoyiyang/Desktop/sector rotation/Fundamentals/EV2EBIT.xlsx', index_col = '3 Months Ending').drop("Unnamed: 0", axis = 1)
df_EV2EBITDA = pd.read_excel('/Users/Shuoyiyang/Desktop/sector rotation/Fundamentals/EV2EBITDA.xlsx', index_col = '3 Months Ending').drop("Unnamed: 0", axis = 1)
df_DivYield = pd.read_excel('/Users/Shuoyiyang/Desktop/sector rotation/Fundamentals/Dividend_Yield.xlsx', index_col = '3 Months Ending').drop("Unnamed: 0", axis = 1)
```

```python
import matplotlib.pyplot as plt
df_PE = df_PE.replace("", np.NaN).backfill()
index_list = []
for index in df_PE.columns:
    index_list.append(index)
plt.figure(figsize = (20,10))
plt.xlabel("Time by Quarters")
plt.ylabel("PE")
plt.ylim(0,100)
plt.plot(df_PE)
plt.legend(index_list,loc=2)
```
